```python
import os
import fitz  # PyMuPDF #you may need to install PyMuPDF to use this

# Function to convert a PDF to text
def convert_pdf_to_text(pdf_path, text_path):
    try:
        pdf_document = fitz.open(pdf_path)
        text = ""
        for page_number in range(pdf_document.page_count):
            page = pdf_document.load_page(page_number)
            text += page.get_text()
        with open(text_path, 'w', encoding='utf-8') as text_file:
            text_file.write(text)
        print(f"Converted {pdf_path} to {text_path}")
    except Exception as e:
        print(f"Error converting {pdf_path}: {str(e)}")

# Folder containing PDF files (including subfolders)
folder_path = r'C:\Users\Prath\Documents\Annual reports'#update to the high level folder where you store your annual reports

# Create a folder named "text files" if it doesn't exist, the next two lines create a new folder in the high level folder for the text files to be stored 
text_folder = os.path.join(folder_path, "text files")
os.makedirs(text_folder, exist_ok=True)

# Loop through all files and subfolders in the specified folder, saves your text files 
for root, _, files in os.walk(folder_path):
    for file in files:
        if file.endswith('.pdf'):
            pdf_path = os.path.join(root, file)
            text_path = os.path.join(text_folder, os.path.splitext(file)[0] + '.txt')
            convert_pdf_to_text(pdf_path, text_path)
```

```python
###BASIC PREPROCESSING 
import os
import re

# Function to preprocess text in a file
def preprocess_text_file(input_file_path, output_file_path):
    try:
        with open(input_file_path, 'r', encoding='utf-8') as input_file:
            text = input_file.read().lower()
            
            # Remove non-text characters, except for '.', '!', and newline
            text = re.sub(r'[^a-zA-Z\s.!\n]', '', text)
            
            # Remove numeric characters
            text = re.sub(r'\d+', '', text)
            
            with open(output_file_path, 'w', encoding='utf-8') as output_file:
                output_file.write(text)
        
        print(f"Preprocessed {input_file_path} and saved as {output_file_path}")
    except Exception as e:
        print(f"Error preprocessing {input_file_path}: {str(e)}")

# Folder containing text files
text_folder_path = r'C:\Users\prath\Documents\Annual reports\text files'

# Create a folder named "preprocessed" if it doesn't exist
preprocessed_folder_path = os.path.join(text_folder_path, 'preprocessed')
os.makedirs(preprocessed_folder_path, exist_ok=True)

# Loop through all text files in the specified folder
for filename in os.listdir(text_folder_path):
    if filename.endswith('.txt'):
        input_file_path = os.path.join(text_folder_path, filename)
        output_file_path = os.path.join(preprocessed_folder_path, os.path.splitext(filename)[0] + '_proc.txt')
        preprocess_text_file(input_file_path, output_file_path)
```
